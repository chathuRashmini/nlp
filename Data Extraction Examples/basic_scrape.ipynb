```python
# a package to send requests for web pages
import requests
```

```python
# url = 'http://www.showmeboone.com/sheriff/JailResidents/JailResidents.asp'
url = 'https://www.w3schools.com/html/html_tables.asp'
```

```python
response = requests.get(url)
```

```python
html = response.content # The main content
print(html)
```

```python
# Parse content
from bs4 import BeautifulSoup # A package for extracting markup
soup = BeautifulSoup(html)
print(soup.prettify())  # BeautifulSoup pretty prints the HTML!
```

```python
# find whether there is a <tbody> tag with class name 'stripe'
# table = soup.find('tbody', attrs={'class': 'mdc-data-table__content'})
table = soup.find('table')
print(table.prettify())
```

```python
# Pretty print row by row
for row in table.findAll('tr'):
    print('Row----------------------')
    print(row.text)
```

```python
# Pretty print cell by cell
for row in table.findAll('tr'):
    print()
    print('Row----------------------')
    for cell in row.findAll('td'):
        print('column: ', cell.text)
```

```python
# Remove the annoying non-break-space   
for row in table.findAll('tr'):
    print()
    print('Row----------------------')
    for cell in row.findAll('td'):
        print('column: ' + cell.text.replace('&nbsp;', ''))
```

```python
# Print as a Python List 
for row in table.findAll('tr'):
    list_of_cells = []
    
    for cell in row.findAll('td'):
        text = cell.text.replace('&nbsp;', '')
        list_of_cells.append(text)
    
    print(list_of_cells)
```

```python
 # Make it into a list of lists
list_of_rows = []

for row in table.findAll('tr'):
    list_of_cells = []
    
    for cell in row.findAll('td'):
        text = cell.text.replace('&nbsp;', '')
        list_of_cells.append(text)
    
    list_of_rows.append(list_of_cells)

print(list_of_rows)
```

```python
# to write it as a CSV file and view using spreadsheet
import csv

outfile = open("./inmates.csv", "w")
writer = csv.writer(outfile)
writer.writerows(list_of_rows)
outfile.close()
```
